```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Movie.csv')
df.head()
```

```python
#convert the dataset rows to coulms and vice-versa / convert long format data to wide format.
user_df = df.pivot(index = 'userId', columns = 'movie', values = 'rating').reset_index(drop=True)
```

```python
user_df
```

```python
user_df.index = df.userId.unique()
```

```python
user_df
```

```python
#impute na values with 0, as in the cx has not watched the movie
user_df.fillna(0, inplace = True)
```

```python
user_df
```

```python
#Calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_df.values, metric= 'cosine')
```

```python
user_sim1 = 1 - pairwise_distances(user_df.values, metric= 'correlation')
```

```python
user_sim
```

```python
user_sim1
```

```python
#convert user_sim into a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df1 = pd.DataFrame(user_sim1)
```

```python
#Set index and column names to user id.
user_sim_df.index = df.userId.unique()
user_sim_df.columns = df.userId.unique()
```

```python
user_sim_df
```

```python
#Set index and column names to user id.
user_sim_df1.index = df.userId.unique()
user_sim_df1.columns = df.userId.unique()
```

```python
user_sim_df1
```

```python
#fill diagonal values with 0
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#fill diagonal values with 0
np.fill_diagonal(user_sim1, 0)
user_sim_df1.iloc[0:5, 0:5]
```

```python
#Most similar values in top 5 rows
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#Most similar values in top 5 rows
user_sim_df1.idxmax(axis=1)[0:5]
```
